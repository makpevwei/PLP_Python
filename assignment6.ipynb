```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Load and Explore the Dataset
# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, header=None, names=columns)
```

```python
# Display the first few rows of the dataset
print("First five rows of the dataset:")
df.head()
```

```python
# Check data types and missing values
print("\nData types and missing values:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)  # If there were missing values, this would remove them
```

```python
# Task 2: Basic Data Analysis
# Compute basic statistics
print("\nBasic statistics of numerical columns:")
print(df.describe())
```

```python
# Perform groupings on a categorical column and compute the mean of a numerical column
grouped = df.groupby('species').mean()
print("\nMean values grouped by species:")
print(grouped)
```

```python
# Task 3: Data Visualization
# Line chart showing trends (example with petal_length for demonstration)
df_sorted = df.sort_values('petal_length')
plt.plot(df_sorted['petal_length'], label='Petal Length', color='blue')
plt.title("Petal Length Trends")
plt.xlabel("Index")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```

```python
# Bar chart comparing average petal length per species
sns.barplot(x=grouped.index, y=grouped['petal_length'])
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.show()
```

```python
# Histogram of sepal length
plt.hist(df['sepal_length'], bins=15, color='orange', edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot of sepal length vs. petal length
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, palette="deep")
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title='Species')
plt.show()
```
